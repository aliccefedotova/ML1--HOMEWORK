```python
end_seminar2 = end_seminar2.rename(columns = {'18_group': 'group_18', '17_group': 'group_17'})
```

```python
group_18 = end_seminar2[['group_18','fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended']].copy()
group_18 = group_18[group_18.group_18 > 0]
group_17 = end_seminar2[['group_17','fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended']].copy()
group_17 = group_17[group_17.group_17 > 0]
```
